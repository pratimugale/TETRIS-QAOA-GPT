```python
max_nodes = 8
```

```python
graphs_nx_dict = defaultdict(dict)

for name, elist in elist_dict.items():
    shifted_elist = []
    
    for n1, n2 in elist:
        cur_e_weight = 1.0
        shifted_elist.append((n1+1, n2+1, cur_e_weight))
    graph_nx = nx.from_edgelist(elist)
    n_nodes = graph_nx.number_of_nodes()

    if n_nodes <= max_nodes:
        graphs_nx_dict[name]['elist'] = shifted_elist
        graphs_nx_dict[name]['n_nodes'] = n_nodes
```
