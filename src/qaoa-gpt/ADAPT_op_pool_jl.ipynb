## Opening pool

```julia
import SparseArrays
import ADAPT
import CSV
import DataFrames
import DataFrames: groupby
import Serialization
import LinearAlgebra: norm
import Graphs
import JSON
```

```julia
n_nodes = 12;
```

```julia
pooltype="qaoa_double_pool";
```

```julia
pool = ADAPT.ADAPT_QAOA.QAOApools.qaoa_double_pool(n_nodes);
```

```julia
for i in 1:length(pool)
    println(i, ": ", pool[i][1].pauli)
end
```

```julia
pauli_json_list = []

for i in 1:length(pool)
    pauli_sparse_matrix = SparseArrays.sparse((Matrix(pool[i][1].pauli)))
    A = pauli_sparse_matrix
    sparse_dict = Dict(
        "op_index" => i,
        "pauli_string" => string(pool[i][1].pauli),
        "m" => size(A, 1),
        "rowval" => A.rowval,
        "colptr" => A.colptr,
        "nzval" => A.nzval,
    )
    push!(pauli_json_list, sparse_dict)
end
```

```julia
json_str = JSON.json(pauli_json_list);
```
